```python
#NAME : VS Prudhvi Rajeev Kumar
#REGD NO: 2141011145
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display the first few rows of the DataFrame
print(df.head())

# Find the shape
shape = df.shape
print(f"Number of rows (data points): {shape[0]}")
print(f"Number of columns (features): {shape[1]}")

# Check for null values
null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)

# Find data types of each column
data_types = df.dtypes
print("Data types of each column:\n", data_types)

# Get statistical information for each numerical column
statistics = df.describe()
print("Statistical information for numerical columns:\n", statistics)

# Convert categorical columns to numerical columns using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['species'], drop_first=True)
print("Data after converting categorical columns to numerical columns:\n", df_encoded.head())

# Find the correlation matrix
correlation_matrix = df_encoded.corr()
print("Correlation matrix:\n", correlation_matrix)

# Plot histograms for numerical columns
df_encoded.hist(bins=30, figsize=(20, 15), layout=(3, 3))
plt.suptitle('Histograms of Numerical Columns')
plt.show()

# Box plots for numerical columns
plt.figure(figsize=(12, 8))
df_encoded.boxplot()
plt.title('Box Plot of Numerical Columns')
plt.show()

# Scatter plots to see relationships between different features
plt.figure(figsize=(12, 8))
colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['species'].map(colors))
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.show()

plt.figure(figsize=(12, 8))
plt.scatter(df['sepal width (cm)'], df['petal width (cm)'], c=df['species'].map(colors))
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot of Sepal Width vs Petal Width')
plt.show()
```
